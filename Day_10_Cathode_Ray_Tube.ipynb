```python
marks = [20, 60, 100, 140, 180, 220]
cycle = 1
x = 1
sum = 0
i = 0
in_addx = False
while i < len(lines):
  if cycle in marks:
    sum += cycle*x

  line = lines[i].split(" ")
  if line[0] == "noop":
    i += 1
  elif line[0] == "addx":
    if in_addx:
      i += 1
      x += int(line[1])
      in_addx = False
    else:
      in_addx = True

  pixel_left = x-1
  pixel_right = x+1
  if cycle % 40 >= pixel_left and cycle %40 <= pixel_right:
    print("#",end="")
  else:
    print(".",end="")
  if cycle % 40 == 0:
    print("")

  cycle += 1
sum
```
