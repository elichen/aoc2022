```python
input = '''#.#####
#.....#
#>....#
#.....#
#...v.#
#.....#
#####.#'''
input = input.split("\n")
```

```python
input = '''#.######
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
######.#'''
input = input.split("\n")
```

```python
input = open("input.txt").readlines()
input = [x.rstrip('\n') for x in input]
```

```python
h,w = len(input)-2, len(input[0])-2
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))
```

```python
def read_map():
  global map
  map = np.zeros((h,w),dtype=int)
  for row,line in enumerate(input):
    for col,c in enumerate(line):
      if c == '>': map[row-1,col-1] = 1
      if c == 'v': map[row-1,col-1] = 2
      if c == '<': map[row-1,col-1] = 4
      if c == '^': map[row-1,col-1] = 8
```

```python
def print_map():
  print(f"# {'#'*w}")
  for row in range(h):
    print("#",end="")
    for col in range(w):
      if map[row,col] == 0: print(".",end="")
      if map[row,col] == 1: print(">",end="")
      if map[row,col] == 2: print("v",end="")
      if map[row,col] == 4: print("<",end="")
      if map[row,col] == 8: print("^",end="")
      if bin(map[row,col])[2:].count('1') == 2 : print("2",end="")
      if bin(map[row,col])[2:].count('1') == 3 : print("3",end="")
      if bin(map[row,col])[2:].count('1') == 4 : print("4",end="")
    print("#")
  print(f"{'#'*w} #")
```

```python
from tqdm.notebook import tqdm
```

```python
def run():
  global map
  read_map()
  cur_set = set()
  for t in tqdm(range(100000)):
    # print("t:",t)
    # print_map()
    new_map = np.zeros((h,w),dtype=int)
    for row in range(h):
      for col in range(w):
        if map[row,col] & 1: new_map[row,col+1 if col<w-1 else 0] |= 1
        if map[row,col] & 2: new_map[row+1 if row<h-1 else 0,col] |= 2
        if map[row,col] & 4: new_map[row,col-1 if col>0 else w-1] |= 4
        if map[row,col] & 8: new_map[row-1 if row>0 else h-1,col] |= 8
    map = new_map
    new_set = set()
    for p in cur_set:
      if p == (h-1,w-1): print("GOAL!",t+1); return
      if map[p] == 0: new_set.add(p)
      if p[1]<w-1 and map[p[0],p[1]+1] == 0: new_set.add((p[0],p[1]+1))
      if p[0]<h-1 and map[p[0]+1,p[1]] == 0: new_set.add((p[0]+1,p[1]))
      if p[1]>0 and map[p[0],p[1]-1] == 0: new_set.add((p[0],p[1]-1))
      if p[0]>0 and map[p[0]-1,p[1]] == 0: new_set.add((p[0]-1,p[1]))
    if map[0,0] == 0: new_set.add((0,0))
    # print(new_set)
    cur_set = new_set
run()
print_map()
```

```python
START = (-1,-1,0)
GOAL1 = (-99,-99,1)
START2 = (-1,-1,2)
def run2():
  global map
  read_map()
  cur_set = set([START])
  for t in tqdm(range(100000)):
    # print("t:",t)
    # print_map()
    new_map = np.zeros((h,w),dtype=int)
    for row in range(h):
      for col in range(w):
        if map[row,col] & 1: new_map[row,col+1 if col<w-1 else 0] |= 1
        if map[row,col] & 2: new_map[row+1 if row<h-1 else 0,col] |= 2
        if map[row,col] & 4: new_map[row,col-1 if col>0 else w-1] |= 4
        if map[row,col] & 8: new_map[row-1 if row>0 else h-1,col] |= 8
    map = new_map
    new_set = set()
    for p in cur_set:
      if p == START or p == START2:
        new_set.add(p)
        if map[0,0] == 0: new_set.add((0,0,p[2]))
        continue
      if p == GOAL1:
        new_set.add(p)
        if map[h-1,w-1] == 0: new_set.add((h-1,w-1,p[2]))
        continue
      if p == (h-1,w-1,2): print("GOAL!",t+1); return
      if p == (h-1,w-1,0): new_set.add(GOAL1)
      if p == (0,0,1): new_set.add(START2)
      if map[p[0],p[1]] == 0: new_set.add(p)
      if p[1]<w-1 and map[p[0],p[1]+1] == 0: new_set.add((p[0],p[1]+1,p[2]))
      if p[0]<h-1 and map[p[0]+1,p[1]] == 0: new_set.add((p[0]+1,p[1],p[2]))
      if p[1]>0 and map[p[0],p[1]-1] == 0: new_set.add((p[0],p[1]-1,p[2]))
      if p[0]>0 and map[p[0]-1,p[1]] == 0: new_set.add((p[0]-1,p[1],p[2]))
    # print(new_set)
    cur_set = new_set
run2()
print_map()
```
