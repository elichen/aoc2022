```python
lines = [x.strip() for x in lines]
```

```python
import numpy as np
```

```python
width, height = len(lines[0]), len(lines)
m = np.zeros((height,width))
start = None
end = None
for i in range(len(lines)):
  line = lines[i]
  for j in range(len(line)):
    if line[j] == "S":
      start = (i,j)
      m[i,j] = 0
    elif line[j] == "E":
      end = (i,j)
      m[i,j] = ord('z') - ord('a')
    else:
      m[i,j] = ord(line[j]) - ord('a')
```

```python
def steps(start):
  steps = 0
  positions = [start]
  visited = set([start])
  while len(positions):
    # print(f"steps:{steps}, num visited{len(visited)} positions:{positions}")
    steps += 1
    next_pos = set()
    for pos in positions:
      # print(pos)
      visited.add(pos)
      y,x = pos[0],pos[1]
      neighbors = [(y,x-1),(y,x+1),(y-1,x),(y+1,x)]
      for node in neighbors:
        if node[1] < 0 or node[1] >= width or node[0] < 0 or node[0] >= height:
          continue
        if node in visited:
          continue
        if m[node] > m[pos]+1:
          continue
        if node == end:
          return steps
        next_pos.add(node)
      # print("next_pos",next_pos)
      positions = next_pos
  else:
    return width*height
print(steps(start))
```

```python
min_steps = width*height
for i in range(width):
  for j in range(height):
    if m[j,i] == 0:
      start = (j,i)
      tmp = steps(start)
      print(start,tmp)
      if tmp < min_steps:
        min_steps = tmp
min_steps
```
