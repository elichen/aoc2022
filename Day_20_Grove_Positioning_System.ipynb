```python
mix = [x for x in range(len(lines))]
for i in range(len(lines)):
  for j in range(len(lines)):
    if mix[j] == i:
      mix.pop(j)
      dst = (j+lines[i])%len(mix)
      # if dst == 0 and lines[i] < 0: dst = len(mix)
      # if dst == 0: dst = len(mix)
      mix.insert(dst, i)
      break
for j in range(len(lines)):
  if lines[mix[j]] == 0:
    zero_i = j
    break
sum([lines[mix[(zero_i+x)%len(mix)]] for x in [1000,2000,3000]])
```
