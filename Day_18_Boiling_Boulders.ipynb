```python
lines = '''2,2,2
1,2,2
3,2,2
2,1,2
2,3,2
2,2,1
2,2,3
2,2,4
2,2,6
1,2,5
3,2,5
2,1,5
2,3,5'''
lines = lines.split("\n")
```

```python
lines = open("input.txt").readlines()
```

```python
lines = [x.strip() for x in lines]
len(lines)
```

```python
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))
```

```python
max_dims = (0,0,0)
for line in lines:
  x,y,z = map(int,line.split(","))
  max_dims = (max(x,max_dims[0]), max(y, max_dims[1]), max(z, max_dims[2]))
max_dims
```

```python
m = np.zeros(list(map(lambda x:x+2,max_dims)))
m.shape
```

```python
for line in lines:
  x,y,z = map(int,line.split(","))
  m[x,y,z] = 1
```

```python
total = 0
for x in range(max_dims[0]+1):
  for y in range(max_dims[1]+1):
    for z in range(max_dims[2]+1):
      if m[x,y,z]:
        sides = 6
        if x>0: sides -= m[x-1,y,z]
        if x<max_dims[0]: sides -= m[x+1,y,z]
        if y>0: sides -= m[x,y-1,z]
        if y<max_dims[1]: sides -= m[x,y+1,z]
        if z>0: sides -= m[x,y,z-1]
        if z<max_dims[2]: sides -= m[x,y,z+1]
        # print(x,y,z, sides)
        total += sides
int(total)
```

```python
to_visit = [(0,0,0)]
while len(to_visit):
  next = to_visit.pop()
  if m[next] == 0:
    m[next] = 8
    x,y,z = next
    if x>0: to_visit.append((x-1,y,z))
    if x<m.shape[0]-1: to_visit.append((x+1,y,z))
    if y>0: to_visit.append((x,y-1,z))
    if y<m.shape[1]-1: to_visit.append((x,y+1,z))
    if z>0: to_visit.append((x,y,z-1))
    if z<m.shape[2]-1: to_visit.append((x,y,z+1))
```
