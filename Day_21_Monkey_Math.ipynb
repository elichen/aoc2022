```python
input = '''root: pppw + sjmn
dbpl: 5
cczh: sllz + lgvd
zczc: 2
ptdq: humn - dvpt
dvpt: 3
lfqf: 4
humn: 5
ljgn: 2
sjmn: drzm * dbpl
sllz: 4
pppw: cczh / lfqf
lgvd: ljgn * ptdq
drzm: hmdt - zczc
hmdt: 32'''
input = input.split('\n')
```

```python
input = open("input.txt").readlines()
```

```python
input = [x.strip() for x in input]
```

```python
g = {}
```

```python
import re
```

```python
for line in input:
  arr = re.split(' |:', line)
  if len(arr) > 3:
    g[arr[0]] = {'op':arr[3], 'l':arr[2], 'r':arr[4]}
  else:
    g[arr[0]] = {'num':int(arr[2])}
```

```python
def eval_monkey(node):
  if 'num' in node:
    return node['num']
  else:
    return eval(f"{eval_monkey(g[node['l']])} {node['op']} {eval_monkey(g[node['r']])}")
eval_monkey(g['root'])
```

```python
def mark_human(node_name):
  result = None
  if node_name == 'humn':
    result = True
  elif 'num' in g[node_name]:
    result = False
  else:
    res1 = mark_human(g[node_name]['l'])
    res2 = mark_human(g[node_name]['r'])
    result = res1 or res2
  g[node_name]['contains_humn'] = result
  return result
mark_human('root')
g
```

```python
# reduces human nodes to list of inverse operations
def reduce_pair(node):
  if 'num' in node and node['contains_humn']:
    return []
  if g[node['l']]['contains_humn']:
    if node['op'] == '+':
      return [f"- {eval_monkey(g[node['r']])}"] + reduce_pair(g[node['l']])
    if node['op'] == '-':
      return [f"+ {eval_monkey(g[node['r']])}"] + reduce_pair(g[node['l']])
    if node['op'] == '*':
      return [f"/ {eval_monkey(g[node['r']])}"] + reduce_pair(g[node['l']])
    if node['op'] == '/':
      return [f"* {eval_monkey(g[node['r']])}"] + reduce_pair(g[node['l']])
  if g[node['r']]['contains_humn']:
    if node['op'] == '+':
      return [f"- {eval_monkey(g[node['l']])}"] + reduce_pair(g[node['r']])
    if node['op'] == '-':
      return [f"- {eval_monkey(g[node['l']])}", "* -1"] + reduce_pair(g[node['r']])
    if node['op'] == '*':
      return [f"/ {eval_monkey(g[node['l']])}",] + reduce_pair(g[node['r']])
    if node['op'] == '/':
      return [f"* {eval_monkey(g[node['l']])}", "** -1"] + reduce_pair(g[node['r']])

def reduce_root(node):
  if g[node['l']]["contains_humn"]:
    num_node = 'r'
    humn_node = 'l'
  else:
    num_node = 'l'
    humn_node = 'r'
  num = eval_monkey(g[node[num_node]])
  ops = reduce_pair(g[node[humn_node]])
  for op in ops:
    num = eval(f"{num} {op}")
  return num

reduce_root(g['root'])
```
